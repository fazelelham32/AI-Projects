Min, Max, Mean, STD

```python
a = tf.constant([[1, 2], [3, 4]], dtype=float32)
a
```

```python
a = tf.constant([[1, 2], [3, 4.0]])
a
```

```python
min = tf.reduce_min(a)
min.numpy()
```

```python
min = tf.reduce_min(a, axis=None)
min.numpy()
```

```python
min = tf.reduce_min(a, axis=0)
min.numpy()
```

```python
min = tf.reduce_min(a, axis=1)
min.numpy()
```

```python
max = tf.reduce_max(a, axis=None)
max.numpy()
```

```python
max = tf.reduce_max(a, axis=0)
max.numpy()
```

```python
max = tf.reduce_max(a, axis=1)
max.numpy()
```

```python
avg = tf.reduce_mean(a)
avg
```

```python
avg = tf.reduce_mean(a)
avg.numpy()
```

```python
# numpy
import numpy as np
a = np.array([[1, 2], [3, 4]])
a.mean()
```

```python
avg = tf.reduce_mean(a)
avg.numpy()
```

```python
avg = tf.reduce_mean(a, axis=0)
avg.numpy()
```

```python
avg = tf.reduce_mean(a, axis=1)
avg.numpy()
```

```python
std = tf.math.reduce_std(a)
std.numpy()
```

```python
sum = tf.reduce_sum(a)
sum.numpy()
```
