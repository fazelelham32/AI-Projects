```python
import numpy as np
```

```python
class neural_network:
  def __init__(self, n_input, n_nroun):
    self.weights = 0.01 * np.random.randn(n_input, n_nroun)
    self.biases = 0.01 * np.random.randn(1, n_nroun)

  def Dense(self, z):
    # Transpose the input z to align the shapes for matrix multiplication
    output = np.dot(z, self.weights) + self.biases
    return output

  def sigmoid(self, z):
    #z = w*x + b
    #sigmoid = 1 / (1e**-z)

    return 1 / (1 + np.exp(-z))

  def softmax(self, z, Z):

    #z = w*x + b
    #Z = sigma z
    return np.exp(z) / sum(np.exp(Z))

dense1 = neural_network(2, 4)
x_input = np.array([[0.3, 0.9]])

output1 = dense1.Dense(x_input)
print(output1)

output = dense1.sigmoid(output1)
print(output)
```

```python
dense2 = neural_network(4, 8)
output2 = dense2.Dense(output1)
print(output2)
```

```python
outpu2 = dense2.sigmoid(output2)
print(output2)
```

```python
dense3 = neural_network(8, 2)
output3 = dense3.Dense(output2)
print(output3)
```

```python
Z = 0
for i in output3:
  Z += np.exp(i)

output_softmax = dense3.softmax(output3, Z)
print(output_softmax)
```
