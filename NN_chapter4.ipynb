```python
import numpy as np
```

```python
inputs = np.array([0, 0])
b = np.array([-1.5, -.5])
```

```python
weights = np.array([[1, 1] , [1, 1]])
```

```python
and_or = np.dot(weights, inputs) + b.T
and_or
```

```python
if and_or[0] > 0:
    and_or[0] = 1
else:
   and_or[0] = 0

if and_or[1] > 0:
    and_or[1] = 1
else:
    and_or[1] = 0
```

```python
and_or
```

```python
weights_xor = np.array([-1, 1])
```

```python
xor = np.dot(weights_xor, and_or)
xor
```

```python
if xor > 0:
    xor = 1
else:
    xor = 0
```
